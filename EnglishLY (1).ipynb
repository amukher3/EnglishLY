Following is an effort to extract and label _parts of speech_ tags in english language texts from PDF copies. This project uses 
Spacy to extract and label the various _parts of speech_ tags in the text. This project also tries to find out the phrases like 
_noun phrases_ and _verb phrases_ in english language text. 


```python
import PyPDF2
```

```python
#To run this just change the path to your downloaded copy of the PDF. 
f = open('C:/Users/Abhishek Mukherjee/Downloads/Short-stories-from-100-Selected-Stories.pdf','rb')
pdf_reader = PyPDF2.PdfFileReader(f)
```

```python
pdf_reader.numPages
```

```python
#Here one can change the page number to
#find the POS for a particular page in 
#the PDF. 

page_one = pdf_reader.getPage(10)
```

```python
page_one_text = page_one.extractText()
```

```python
page_one_text
```

```python
import re
myLst=re.split('\n',page_one_text)
print(myLst)
```

```python
str1= "!#$%&'()*+,-./:;?@[\]^_`{|}~"

lst2=[]
for i in str1:
    lst2.append(str(i))
    
finLst=[]
for i in myLst:
    
    if i in lst2 or i==' ' or i=='' or len(i)==1:
        continue
    else:    
        finLst.append(i)
         
for i in finLst:
    print(i)
    
finStr=''.join(finLst)    


finStr2=[]
finStr2=finStr.split('.')
```

```python
finStr2[1]
```

```python
# Import the displaCy library
from spacy import displacy

for i in finStr2:
    print(i)
```

```python
#Different POS and their relative
#positions in a sentence are shown here

import spacy
nlp = spacy.load("en_core_web_sm")

for i in finStr2:
    
    doc = nlp(i)
    
    displacy.render(doc, style='dep', jupyter=True, options={'distance': 110})
    
    for token in doc:
        print(f'{token.text:{10}} {token.pos_:{7}} {token.dep_:{7}} {spacy.explain(token.dep_)}')
```

```python
options = {'distance': 110, 'compact': 'True', 'color': 'yellow', 'bg': '#09a3d5', 'font': 'Times'}
displacy.render(doc, style='dep', options=options)
```

```python
#Noun phrashes and verbs: 

for i in finStr2:
    doc = nlp(i)
    
    # Analyze syntax
    for token in doc:
        if token.pos_ == "VERB":
            if token.lemma_ !=[]:
                print("Verbs:",token.lemma_ )
    [print("Noun phrases:",chunk.text) for chunk in doc.noun_chunks if chunk.text!=[]]
    
```
